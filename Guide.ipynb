# Финальный проект первого семестра DLSchool. 

## Classification telegram bot.

В данном проекте мы хотим сделать телеграм бота, который будет угадывать, что есть у нас на картинке.

В данной проекте у нас есть две папки:
* model - папка для ноутбуков с экспериментами. Именно в этой части проекта мы будем строить свою модель, обучать ее и тестировать.
* telegram_bot - папка с .py файлами, которые относятся к продакшн части нашего проекта. Здесь будет лежать наш файл с телеграм ботом, а так же перенесенная из тетрадки модель, вся работа которой скрыта в классе.

В бейзлайне используется библиотека для ботов python telegram bot. Документация к ней легко гуглится. 

Если при запуске бейзлайна вам не хватает какой-то библиотеки - просто забивайте в гугл название отсутствующего модуля и pip'ом устанавливайте нужную библиотеку.

Думаю, что наша работа будет максимально продуктивной, если вы будете двигаться по следющему курсу:

0. Написать BotFather и получить свой токен. В папке telegram_bot создать файл telegram_token.py с единственной строчкой: token = 'ВАШ ТОКЕН ЗДЕСЬ'. **Будьте осторожны и не загрузите свой токен на гитхаб!!!** Я уже включил файл telegram_token.py в .gitignore, поэтому токен не попадет на гитхаб, если вы будете держать его только в этом файле. **ВАЖНО: чтобы .gitignore сработал его надо добавить первым, то есть перед добавлением всех остальных файлов вызвать команду git add .gitignore из корня проекта**.
1. Созддайте свой **гитхаб репозиторий** и сделайте начальный коммит, состоящий из файлов бейзлайна, скиньте ссылку в слак, чтобы я имел доступ к репозиторию. Эту тетрадку включать не стоит, для удобства я внесу её в .gitignore. Если вы не можете в гит, то напишите мне, я помогу разобраться.
2. Выберите датасет и, опять же, напишите о своем выборе в слаке. Если нужно, обработайте его и приведите к виду, в котором его будет удобно обрабатывать. 
3. Обучите сеть и улучшите ее. Для этого Вы можете использовать fast ai. Вы получите много пользы от проекта, если в процессе изучите основные возможности этой библиотеки (также можно прочитать статьи про методы, которые вы используете из библиотеки, если у вас останется на это время).
4. Перенесите модель в отдельный класс в папке telegram_bot. По сути нужно просто загрузить веса.
5. Доработайте самого **бота**: добавьте приветствие и объяснение принципа работы. Будет хорошо, если все строчки, которые отправляет бот вы вынесете в отдельный файл. Например, config.py.
6. Теперь все готово и остались только **косметические изменения**: добавление docstring, исправение кодстайла, рефакторинг и т.д. 
7. ...
8. PROFIT!!!

Возможно, вы не успеете сделать все, но постарайтесь выполнить как можно больше пунктов. В то же время не следует ограничиваться этими пунктами, если Вы видите какой-то способ сделать свою работу еще лучше.
